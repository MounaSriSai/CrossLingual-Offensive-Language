```python

import pandas as offpn
import imblearn
from imblearn.over_sampling import RandomOverSampler as off_langOvrsamp
from collections import Counter as off_langCnt
from sklearn.feature_extraction.text import TfidfVectorizer as off_langTF
from sklearn.decomposition import PCA as off_langPCA

off_data = 'processedArabic.csv'
off_lang = offpn.read_csv(off_data, encoding='latin1')
print("\nTarget Variants:\n",off_lang['Class'].value_counts())

off_langXin_Arab = off_lang.drop('Class', axis=1)
off_langYop_Arab = off_lang['Class']

print(off_langXin_Arab)
print(off_langYop_Arab)

off_langVc = off_langTF(analyzer='word', ngram_range= (1,1))
off_langXin_Arab = off_langVc.fit_transform(off_lang['Tweet'].values.astype('U')).toarray()

off_langPcaM = off_langPCA(n_components = 5)
off_langXin_Arab = off_langPcaM.fit_transform(off_langXin_Arab)
print("\nNumber of ARABIC Components:",off_langXin_Arab.shape)

print('\nProcessed Arabic Data :  %s' % off_langCnt(off_langYop_Arab))
off_langM = off_langOvrsamp()
off_langXin_Arab, off_langYop_Arab = off_langM.fit_resample(off_langXin_Arab, off_langYop_Arab)
print('Oversampled Arabic Data  : %s' % off_langCnt(off_langYop_Arab))
```

```python
off_data = 'processedEnglish.csv'
off_lang = offpn.read_csv(off_data, encoding='latin1')
print("\nTarget Variants:\n",off_lang['class'].value_counts())

off_langXin_Eng = off_lang.drop('class', axis=1)
off_langYop_Eng = off_lang['class']

print("\nEnglish Data without vectorization:",off_lang.shape)
off_langVc = off_langTF(analyzer='word', ngram_range= (1,1))
off_langXin_Eng = off_langVc.fit_transform(off_lang['tweet'].values.astype('U')).toarray()
print("English Data after vectorization:",off_langXin_Eng.shape)

off_langPcaM = off_langPCA(n_components = 5)
off_langXin_Eng = off_langPcaM.fit_transform(off_langXin_Eng)
print("\nNumber of ENGLISH Components", off_langXin_Eng.shape)

print('\nProcessed English Data    :  %s' % off_langCnt(off_langYop_Eng))
off_langM = off_langOvrsamp()
off_langXin_Eng, off_langYop_Eng = off_langM.fit_resample(off_langXin_Eng, off_langYop_Eng)
print('Oversampled English Data  : %s' % off_langCnt(off_langYop_Eng))
```
